```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/angelesmarin/Desktop/hulu_data.csv")
```

### handle missing values 

```python
df.dropna(inplace=True) #drop columns & rows with na
```

### remove duplicates 

```python
df.drop_duplicates(inplace=True) #drop columns & rows with duplicates
```

### fix data types 

```python
df['releaseYear'] = pd.to_numeric(df['releaseYear'], errors='coerce')
```

```python
df['imdbAverageRating'] = pd.to_numeric(df['imdbAverageRating'], errors='coerce')
```

```python
df['imdbNumVotes'] = pd.to_numeric(df['imdbNumVotes'], errors='coerce')
```

```python
# ^numeric, if cant -> nan
```

```python
df['imdbId'] = df['imdbId'].str.strip()  #remove leading/trailing spaces
```

```python
df['title'] = df['title'].str.strip()
```

```python
df['type'] = df['type'].str.strip()
```

```python
df['genres'] = df['genres'].str.strip()
```

```python
df['availableCountries'] = df['availableCountries'].str.strip()
```
